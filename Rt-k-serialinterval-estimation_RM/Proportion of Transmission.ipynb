```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp = pd.read_excel('transprop_curve_final.xlsx',sheet_name="graph_confirmedtraced_956")
```

```python
plt.figure(figsize=(7,6))
plt.plot(temp['Proportion of infectious cases (Ranked by infectiousness)'],temp['Proportion of onward transmission'],'-',color='steelblue',linewidth=2.5,label='Proportion of onward transmission (%)')
plt.plot(temp['Proportion of infectious cases (Ranked by infectiousness)'],temp['Proportion of contacts'],'--',color='steelblue',linewidth=2.5,label= 'Proportion of contacts (%)')
plt.xlabel("Proportion of infectious cases ranked by infectiousness (%)")
plt.legend(frameon=False)
plt.grid(alpha=0.2)
plt.savefig('Figures/prop956.png',dpi=300)
plt.show()
```
